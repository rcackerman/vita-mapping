```python
import pandas as pd
```

### Importing 2019 ACS dataset

```python
df2019 = pd.read_csv("ACS 2019_income_data.csv")
```

```python
df2019.sample(5)
```

```python
df2019['vita_eligible'] = df2019.less_than_10000 + \
                          df2019.total_10000_to_14999 + \
                          df2019.total_15000_to_19999 + \
                          df2019.total_20000_to_24999 + \
                          df2019.total_25000_to_29999 + \
                          df2019.total_30000_to_34999 + \
                          df2019.total_35000_to_39999 + \
                          df2019.total_40000_to_44999 + \
                          df2019.total_45000_to_49999 + \
                          df2019.total_50000_to_59999
```

```python
df2019['vita_per_capita'] = df2019.vita_eligible / df2019.total
```

```python
df2019['ZIP'] = df2019['Geographic Area Name'].str.replace('ZCTA5 ', '')
```

Looking at the counts, it appears that there's a duplicated row: zip code 31905.

```python
df2019.ZIP.describe()
```

Looking at the rows for 31905, they look the same.

```python
df2019[df2019.ZIP == '31905']
```

Creating a dataframe with duplicate rows dropped, then comparing the counts and main descriptive statistics of the original and new dataframes to make sure we didn't drop more rows than we expected.

```python
df2019_dedup = df2019.drop_duplicates()
```

```python
df2019.describe()
```

```python
df2019_dedup.describe()
```

```python
df2019_dedup[['ZIP', 'total', 'me_total', 'vita_eligible', 'vita_per_capita', 'id']].to_csv("ACS 2019_income_data_zip_accessible.csv", index=False)
```
