```python
import pandas as pd
```

```python
mft_df = pd.read_csv("./data/mft_at_county_level.csv")
vita_df = pd.read_csv("./data/vita_sites_by_county.csv")
county_names = pd.read_csv("./data/acs_household_income_county.csv", dtype={'Full FIPS Code': 'string'})[
    ['acs_state_fips', 'acs_county_fips', 'county_name', 'state_name', 'Full FIPS Code']
]
```

```python
vita_df.info()
```

```python
mft_df.info()
```

```python
mft_df['Any e-filed returns'] = mft_df['#e-filed returns'].apply(lambda x: 'Yes' if (x > 0) else None)
mft_df.dropna(subset=['Any e-filed returns'], inplace=True)
```

```python
mft_vita_df = pd.merge(
    pd.merge(
        county_names, mft_df,
        left_on=['acs_state_fips', 'acs_county_fips'], right_on=['STATE', 'COUNTY'],
        how='left'
    ), vita_df,
    left_on=['acs_state_fips', 'acs_county_fips'], right_on=['STATEFP', 'COUNTYFP'],
    how='left')
```

```python
mft_vita_df.info()
```

```python
mft_vita_df.drop(columns=['acs_state_fips', 'acs_county_fips', 'STATE', 'COUNTY', 'AFFGEOID', 'GEOID'], inplace=True)
```

```python
mft_vita_df.sample(10)
```

```python
import csv
mft_vita_df.to_csv('./data/mft_vita_merged.csv', index=False, quoting=csv.QUOTE_NONNUMERIC)
```
