```python
import pandas as pd
```

```python
vita_sites_df = pd.read_csv("./vita_providers_created_by_our_irs_vita_site_downloader_2022-03-24T15 48 05.984215Z.csv")
```

```python
import geopandas
```

VITA sites dataframe uses WKB coordinates - we want lat/long for Tableau.

```python
vita_sites_df["lat_long"] = geopandas.GeoSeries.from_wkb(vita_sites_df.coordinates)
vita_sites_df["longitude"] = vita_sites_df.lat_long.apply(lambda point: point.x)
vita_sites_df["latitude"] = vita_sites_df.lat_long.apply(lambda point: point.y)
```

Per CFA email, we want only sites where `archived=False`.

```python
vita_sites_df = vita_sites_df[vita_sites_df.archived == False]
```
